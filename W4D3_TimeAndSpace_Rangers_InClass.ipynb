## Time and Space Complexity

#### What is it?

Time and space complexity is the measure of how much time a given action(function) will take to solve a problem. In the same fashion, we determine how much a given data structure will need in terms of memory allocation. A problem can have multiple solutions and finding the optimal solution for the problem needs to be analyzed in time and space.

#### How do we measure Time and Space Complexity?

In order to measure time and space complexity we use Asymptotic analysis. The reason for this is because we need a way to measure different algorithms (functions) based on the size of their inputs in a mathmatical way. For example, we could have a function that is computed as f(n) and another that is g(n^2). All things around the function staying constant, the only thing that changes is the size of the input. ## Stacks and Queues

** Stacks ** as the name suggests is a data structure that allows for data to follow the Last In First Out priciple(LIFO). Think of a stack of pancakes for example. To get the first pancake you would  start with the top and go down.

##### Searching through a stack will be Linear Time O(n) - Constant Space O(1)
##### Selecting the last item will be done in Constant Time O(1) - Constant Space O(1)
##### Adding to the stack should take Constant Time O(1) - Constant Space O(1)

** Queues ** are similar but in this case follow the First In First Out principle(FIFO). Think of this as a line in a black friday sale. The first person camped out for the big screen tv is the first to get it.

##### Searching through a queue will be Linear Time O(n) - Constant Space O(1)
##### Selecting the first item will be done in Constant Time O(1) - Constant Space O(1)
##### Adding to the queue should take Constant Time O(1) - Constant Space O(1)

```python
# Stacks
stack = ['pancake1']
print(stack)
stack.append("pancakce2")
print(stack)
stack.append('pancake3')
print(stack)
stack.append('pancake4')
print(stack)
print(stack.pop())
print(stack)
print(stack.pop())
print(stack)
print(stack.pop())
print(stack)
```

## Linked List (Data Structure)

A linked list is created by using the node class. We create a Node object and create another class to use this node object. We pass the appropriate values thorugh the node object to point the to the next data elements.

There are some advantages and disadvantages with this data structure. **Advantages** Linked Lists can save memory because they can be flexibile with memory management which saves memory. **Disadvantages** Finding or adding to the list requires traversing the entire list.

```python
# Made of Two Classes 
# Node class
# Link_List Class

# Linked_List class
# --traversal
# --insertion
# --delete

# Node Class
# --value -- the actual value of the node
# --next -- points to the next node's value

# A complete implementation of a linked list

# 2 components to the structure
# Node class
# Linked List Class
# node class

#Node class
class Node():
    def __init__(self, value):
        self.value = value
        self.next = None
        
#LinkedList class
class LinkedList():
    def __init__(self):
        self.head = None
        
    #pushOn - place item at the head of the list -- the beginning
    def pushOn(self, new_value):
        new_node = Node(new_value)
        new_node.next = self.head
        self.head = new_node
        
    #insertAfter - place a node after another in the linked list
#     def insertAfter(self, prev_node, new_value):
#         #check if the previous node exists
#         if prev_node is None:
#             print("The given node must not be empty")
#             return
#         #if the prev_node is not empty, we create a new node
#         new_node = Node(new_value)
        
#         # update the new node's next pointer to point to the prev_node's next
#         new_node.next = prev_node.next
        
#         #update pre_node's next to new_node
#         prev_node.next = new_node


    def insertKnownPosition(self, position, new_value):
        
        #creating the new node
        new_node = Node(new_value)
        
        #starting at the beginning of the linked list
        current_node = self.head
        next_node = current_node.next
        
        #traversing linked list until we get to our position we want to insert
        for i in range(position - 1):
            current_node = next_node
            next_node = current_node.next
            
        new_node.next = current_node.next
        current_node.next = new_node 
        
        
    def deleteNode(self, value):
        
        #starting at the beginningh of the linked list
        current_node = self.head
        next_node = current_node.next
        
        #traversing until we get to the node that is the same as our value
        while current_node.value != value and next_node.next:
            print(current_node.value)
            prev_node = current_node
            current_node = next_node
            next_node = current_node.next 
               
        #if we've reached the end of our linked list and value was not found
        if not next_node.next and current_node.value != value:
            return "Value does not exist"
        
        #setting the previous nodes next to the next node and delete our current node 
        prev_node.next = next_node 
        del current_node
        
        
    #append(add) a new node to the END of the linked list
    def append_(self, new_value):
        #create our new node
        new_node = Node(new_value)
        
        #Check if the linked list is empty
        #if it is, make this new node the head node(beginning of the list)
        if self.head is None:
            self.head = new_node
            
        #if not empty, traverse our linked list and append to the end
        last = self.head
        
        ##change current last to the new value, and continuously check for an empty next
        while last.next:
            last = last.next
        
        
        #change the current last node next value to point to the new value
        last.next = new_node
        
        
    #traverse - move through the linked list
    def traverse(self):
        temp = self.head
        #while temp is NOT NONE --> keep on truckin through the list until it is none
        while temp:
            print(temp.value)
            temp = temp.next
        print("end of list")
            
```

```python
# 2.https://www.codewars.com/kata/576bb71bbbcf0951d5000044

def count_positives_sum_negatives(arr):
    if not arr:
        return []
    
    positive_num_count = 0 #constant time and space complexity O(1) 
    negative_num_sum = 0   #constant time and space complexity O(1)
    
    for num in arr: # 'num' iterations where 'num' is the length of the input array
        if num > 0: # This statement takes constant time and space O(1), Since it simply check a single condition.
            positive_num_count += 1 
        elif num < 0:              #constant time and space O(1)
            negative_num_sum += num
            
    return[positive_num_count, negative_num_sum] #constant time and space O(1)

"""
Time Complexity:
The time complexity of this code is O(n), where 'n' is the length of the input array 'arr'. 
The for loop runs 'n' times, where 'n' is the number of elements in the 'arr' list. 
The operations inside the loop (incrementing counts and summing negative numbers) are constant time operations, denoted as O(1).

Space Complexity:
The space complexity is O(1) since the additional space used in the algorithm remains constant, 
regardless of the size of the input array. The variables 'positive_num_count' and 'negative_num_sum' are fixed in size 
and do not scale with the input.

In summary, the time complexity is O(n) because the algorithm's time taken increases linearly with the size of the input array, 
and the space complexity is O(1) because the amount of memory used remains constant regardless of the input size.

"""
```
